```python
import torch
```

```python
bigram_list = {}
characters = []
for name in names:
  name = ['<!>'] + list(name) + ['<!>']
  for ch1, ch2 in zip(name, name[1:]):
    bigram = (ch1, ch2)
    bigram_list[bigram] = bigram_list.get(bigram, 0) + 1
    if ch1 not in characters:
      characters.append(ch1)
    if ch2 not in characters:
      characters.append(ch2)
```

```python
sorted(bigram_list.items(), key = lambda kv: -kv[1])
```

```python
char2idx = {ch: idx for idx, ch in enumerate(characters) }
char2idx
```

```python
idx2char = {idx: ch for ch, idx in char2idx.items()}
idx2char
```

```python
bigram_matrix = torch.zeros(len(char2idx), len(char2idx), dtype = torch.int32)
```

```python
for key, value in bigram_list.items():
  bigram_matrix[char2idx[key[0]]][char2idx[key[1]]] = value
```

```python
bigram_matrix[:2]
```

```python
import plotly.express as px
fig = px.imshow(bigram_matrix.numpy(), text_auto=True, x = list(idx2char.values()), y = list(idx2char.values()))
fig.update_xaxes(side = 'top')
fig.update_layout(height = 1000)
fig.show()
```

```python
probs_matrix = (bigram_matrix + 1).float()
probs_matrix.shape, probs_matrix.sum(1, keepdims = True).shape
```

```python
def generate():
  char = '<!>'
  name = ''
  while True:
      probs = probs_matrix[char2idx[char]]
      char = idx2char[torch.multinomial(probs, num_samples = 1, replacement = True).item()]
      if char == '<!>':
        break
      name += char
  return name
```

```python
log_likelihood = 0
for name in names:
  name = ['<!>'] + list(name) + ['<!>']
  for ch1, ch2 in zip(name, name[1:]):
    probs = probs_matrix[char2idx[ch1]][char2idx[ch2]]
    log_likelihood += torch.log(probs)
```
